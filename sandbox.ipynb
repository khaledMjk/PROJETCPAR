````python
from math import sqrt
def func(t_nbpoints) :
    sqrtNbPoints = int(sqrt(t_nbpoints))
    nbPointsY = t_nbpoints//sqrtNbPoints
    nbPointsX = sqrtNbPoints + (1 if(t_nbpoints%sqrtNbPoints > 0 ) else 0)
    return nbPointsX*nbPointsY
````

````python
t_nbpoints = 120000
nbPoints = func(t_nbpoints)
nbp = 14
nbPoints_loc = []
for _ in range(nbp-1) :
    nbPoints_loc.append(nbPoints//(nbp -1))

for _ in range(nbPoints%(nbp -1)) :
    nbPoints_loc[_] +=1

print(nbPoints_loc)

print([func(_) for _ in nbPoints_loc])

print("{:2f}%".format(((sum([func(_) for _ in nbPoints_loc]) - nbPoints)/nbPoints)*100))
````
